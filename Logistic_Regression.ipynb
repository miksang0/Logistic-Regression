```python
import pandas as pd
path = '/content/Heart.csv'
df = pd.read_csv(path)
df
```

```python

#converting non-numeric values into numeric values.
df['ChestPain'] = df['ChestPain'].astype('category')
df['ChestPain'] = df['ChestPain'].cat.codes
df
```

```python
df['Thal'] = df['Thal'].astype('category')
df['Thal'] = df['Thal'].cat.codes

df['AHD'] = df['AHD'].astype('category')
df['AHD'] = df['AHD'].cat.codes
df
```

```python
#checking null values
df.isnull().sum()
```

```python
#dropping null values
df = df.dropna()
df
```

```python
X = df.drop(columns = 'AHD')
X
```

```python
# dropping Unnamed: 0 from X

X = X.drop(columns = ['Unnamed: 0'])
X
```

```python
y = df['AHD']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=21) #30% for testing 70% for training
X_train
```

```python
X_test
```

```python
#as the number are vastly different, we are using a standard scaler. This helps in the training because no data can be powering during the trainging.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
```

```python
X_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state= 0).fit(X_train_scaled, y_train)
```

```python
log_reg.predict(X_train_scaled)
```

```python
#checking the score
log_reg.score(X_train_scaled, y_train)
```

```python
log_reg.score(X_test_scaled, y_test)
```

```python
log_reg1 = LogisticRegression(random_state= 0,
                              C = 1,
                              fit_intercept = True,
                              ).fit(X_train_scaled, y_train)
```

```python
log_reg1.score(X_train_scaled, y_train)
```

```python
log_reg1.score(X_test_scaled, y_test)
```
